# Definición y Aplciación de Pseudo-Inversa de Moore y Penrose

## Importar Datos

```python
import pandas as pd
data = pd.read_excel("Lanzamientos.xlsx")
data
```

```python
import plotly.express as px
N = data.shape[0]
n_val = np.arange(1, N+1)
pc = np.cumsum(data.values)/n_val
r = np.zeros((2,len(pc)))
r[0,:] = pc
r[1,:] = 1-pc
fig = px.line(r.T, title='Series Observadas', template='plotly_white')
fig
```

## Funcion de Pseudo-Inversa por Descomposicion de Valores Singulares (DVS)

```python
import numpy as np
def pseudoinversa(xt):
    U, S, Vt  = np.linalg.svd(xt,full_matrices=False)
    xt_pinv = Vt.T @ np.diag(1/S) @ U.T
    return xt_pinv
```

## Aplicacion a Datos

```python
N_en = 300
A = r[:,1:N_en+1]@pseudoinversa(r[:,:N_en])
print(A)
```

## Resultados

```python
px.imshow(A,title='Matriz de Transición de Espacio de Estados',text_auto='.4f',color_continuous_scale="greys", template='plotly_white')
```

```python
import plotly.graph_objects as go
xt1 = r[:,:-1].T@A.T

fig = px.line(r.T, title='Proyecciones t+1 o One-step Ahead', template='plotly_white')
[fig.add_trace(go.Scatter(y = xt1[:,i], line={'dash':'dash'})) for i in range (xt1.shape[1])]
fig.add_vline(x=N_en, line_dash='dash')
fig
```
